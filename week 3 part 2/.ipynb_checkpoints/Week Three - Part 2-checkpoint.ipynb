Import required libraries

```python
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms.bipartite as bipartite
import pandas as pd
```

Loading data

```python
G = nx.davis_southern_women_graph()
women = G.graph["top"]
clubs = G.graph["bottom"]
```

Plot the graph

Set up the colors for each node. The women node is in Green and the even node is is Orange

```python
colors = ['Green' if node in women else 'Orange' for node in G.nodes]
```

```python
plt.figsize = (20,20)
nx.draw_shell(G,with_labels=True, node_color = colors)
plt.show()
```

### Network of women

Project bipartite graph onto women nodes. The edges are wighted by the number of friends attending the same events

```python
# project bipartite graph onto women nodes keeping number of co-occurence
W = bipartite.weighted_projected_graph(G, women)
```

Create a summary of the centralities of the graph for the women

```python
#Degree, Eigenvector and Betweenness are weighted by the total number of shared contacts
#DEGREE_CENTRALITY and CLOSENESS are not weighted
summarys = pd.DataFrame(dict(
    DEGREE = dict(W.degree(weight='weight')),
    DEGREE_CENTRALITY = nx.degree_centrality(W),
    EIGENVECTOR = nx.eigenvector_centrality(W, weight='weight'),
    CLOSENESS = nx.closeness_centrality(W),
    BETWEENNESS = nx.betweenness_centrality(W, weight='weight')
)) 
summarys = summarys.sort_values(by = 'DEGREE', ascending = False)
summarys
```

Theresa Anderson and Evelyn Jefferson are the most popular persons (high degree) in the network. They also have the strongest influence (high eigenvector and high closeness) <br>
Olivia Carleton and Flora Price are not so active (low degree) in the network. They mostly function as "brideges" (high betweenness) in the network

```python
nx.draw(W,with_labels=True)
```

### Network of events

Project bipartite graph onto clubs nodes. The edges are wighted by the number of common attendees

```python
# project bipartite graph onto clubs nodes keeping number of common attendees
C = bipartite.weighted_projected_graph(G, clubs)
```

Create a summary of the centralities of the graph for the events

```python
#Degree, Eigenvector and Betweenness are weighted by the total number of shared contacts
#DEGREE_CENTRALITY and CLOSENESS are not weighted
summarys = pd.DataFrame(dict(
    DEGREE = dict(C.degree(weight='weight')),
    DEGREE_CENTRALITY = nx.degree_centrality(C),
    EIGENVECTOR = nx.eigenvector_centrality(C, weight='weight'),
    CLOSENESS = nx.closeness_centrality(C),
    BETWEENNESS = nx.betweenness_centrality(C, weight='weight')
)) 
summarys = summarys.sort_values(by = 'DEGREE', ascending = False)
summarys
```

E8 is the most popular event (high degree) to the persons in the network. Also, it is the key event (high eigenvector and high closeness) that can spread information to all other events. <br>
E6 is not popular as E8. However, it is served (high betweenness) as a bridge to pass information from one set of close events to another set of close events

```python
nx.draw(C,with_labels=True)
```
